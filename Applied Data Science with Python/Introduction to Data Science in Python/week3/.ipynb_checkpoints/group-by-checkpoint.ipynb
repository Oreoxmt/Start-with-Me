````python
df = df.set_index("STNAME")
def fun(item):
    if item[0] < "M":
        return 0
    if item[0] > "Q":
        return 1
    return 2
for group, frame in df.groupby(fun):
    print("There are " + str(len(frame)) + " records in group " + str(group) + " for processing.")
````
