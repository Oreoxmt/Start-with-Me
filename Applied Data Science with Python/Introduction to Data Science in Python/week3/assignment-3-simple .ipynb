```python
import pandas as pd
import numpy as np
```

```python
def answer_one():
    energy = pd.read_excel("Energy+Indicators.xls")
    energy = energy.iloc[17:244,2:]
    energy.columns = ["Country", "Energy Supply", "Energy Supply per Capita", "% Renewable"]
    energy["Energy Supply"] *= 1000000
    energy.where(energy != "...", np.NaN)
    rename_dict = {"Republic of Korea" : "South Korea", "United States of America" : "United States", "United Kingdom of Great Britain and Northern Ireland" : "United Kingdom", "China, Hong Kong Special Administrative Region" : "Hong Kong"}
    energy.Country = energy.Country.replace(rename_dict)
    energy.Country = energy.Country.str.replace(r" \(.*\)", "")
    energy.Country = energy.Country.str.replace(r"[0-9]", "")
  
    GDP = pd.read_csv("world_bank.csv", skiprows = 4)
    GDP["Country Name"] = GDP["Country Name"].replace("Korea, Rep.", "South Korea")
    GDP["Country Name"] = GDP["Country Name"].replace("Iran, Islamic Rep.", "Iran")
    GDP["Country Name"] = GDP["Country Name"].replace("Hong Kong SAR, China", "Hong Kong")
    GDP = GDP[["Country Name", '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']]
    GDP.columns = ["Country", '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']
    print(GDP.shape)
    ScimEn = pd.read_excel("scimagojr-3.xlsx")
    ScimEn = ScimEn[:15]
    print(ScimEn.shape)
    df = pd.merge(ScimEn, GDP, on = "Country")
    df = pd.merge(df, energy, on = "Country")
    df = df.set_index("Country")
    print(df.shape)
    return df
answer_one()
```
