# 中文分词

```python
with open("resources/i-can-eat-glass.txt", encoding = "utf-8") as f:
    text = f.read()
print("Text:\n{}\nText length:{}".format(text, len(text)))
```

## jieba
### 分词 jieba.cut()

```python
import jieba
jieba_cut_result = jieba.cut(text)
print("jieba.cut() returns:{}".format(jieba_cut_result))
```

```python
for word in jieba_cut_result:
    print(word)
    break
```

```python
print("jieba.lcut() returns:{}".format(type(jieba.lcut(text))))
```

```python
jieba_cut_list = list(jieba_cut_result)
#jieba_cut_list = jieba_lcut(text)
print("/".join(jieba_cut_list))
```

```python
results = [
    ("默认", jieba.cut(text)),
    ("全模式", jieba.cut(text, cut_all = True)),
    ("关闭 HMM", jieba.cut(text, HMM = False)),
    ("paddle", jieba.cut(text, use_paddle = True)),
    ("搜索引擎", jieba.cut_for_search(text))
]
for r in results:
    print("{}: {}\n".format(r[0], '/'.join(list(r[1]))))
```

jieba.add_word() 词频 要达到一定值才能达到想要的分词效果

```python
jieba.load_userdict("dict1.txt")
print("/".join(jieba.lcut(text, HMM = True)))
```

```python
jieba.add_word("项目", 100000)
jieba.add_word("地道", 1000)
print("/".join(jieba.lcut(text, HMM = True)))
```
