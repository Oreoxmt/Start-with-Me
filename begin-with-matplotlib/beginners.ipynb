```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X = np.linspace(0, 4*np.pi, 1000)
Y = np.sin(X)
plt.plot(X, Y)
```

```python
X = np.random.uniform(0, 1, 100)
Y = np.random.uniform(0, 1, 100)
plt.scatter(X, Y)
```

```python
X = np.arange(10)
Y = np.random.uniform(1, 10, 10)
plt.bar(X, Y)
```

```python
Z = np.random.uniform(0, 1, (8, 8))
plt.imshow(Z)
```

```python
Z = np.random.uniform(0, 1, (8, 8))
plt.contourf(Z)
```

```python
Z = np.random.uniform(0, 1, 4)
plt.pie(Z)
```

```python
Z = np.random.normal(0, 1, 100)
plt.hist(Z)
```

```python
X = np.arange(5)
Y = np.random.uniform(0, 1, 5)
plt.errorbar(X, Y, Y/4)
```

```python
Z = np.random.normal(0, 1, (100, 3))
plt.boxplot(Z)
```

```python
X = np.linspace(0, 10, 100)
Y = np.sin(X)
plt.plot(X, Y, color = "black")
```

```python
X = np.linspace(0, 10, 100)
Y = np.sin(X)
plt.plot(X, Y, linestyle = "--")
```

```python
X = np.linspace(0, 10, 100)
Y = np.sin(X)
plt.plot(X, Y, linewidth = 5)
```

```python
X = np.linspace(0, 10, 100)
Y = np.sin(X)
plt.plot(X, Y, marker = 'o')
```

```python
X = np.linspace(0, 10, 100)
Y1, Y2 = np.sin(X), np.cos(X)
plt.plot(X, Y1, Y2)
```

```python
X = np.linspace(0, 10, 100)
Y1, Y2 = np.sin(X), np.cos(X)
plt.subplot(2, 1, 1)
plt.plot(X, Y1, color = "C1")
plt.subplot(2, 1, 2)
plt.plot(X, Y2, color = "C0")
```

```python
X = np.linspace(0, 10, 100)
Y1, Y2 = np.sin(X), np.cos(X)
plt.subplot(1, 2, 1)
plt.plot(Y1, X, color = "C1")
plt.subplot(1, 2, 2)
plt.plot(Y2, X, color = "C0")
```
